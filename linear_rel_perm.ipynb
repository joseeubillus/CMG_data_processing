```python
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
# Water saturation 
sw = np.arange(0.2, 1.01, 0.01)
so = 1 - sw
krw = 5/4*(sw) -1/4
krg = -5/4*sw + 5/4

plt.plot(sw, krw, label='krw')
plt.plot(sw, krg, label='krg')
plt.xlabel('Water saturation')
plt.ylabel('Relative permeability')
plt.legend()
plt.gca().margins(y=0)
plt.show()
```

```python
# Save table as txt file 
data = np.vstack((sw, krw, np.flip(krg))).T
np.savetxt('krw.txt', np.round(data,2), delimiter=',',fmt=['%.2f','%.2f','%.2f'])
```
