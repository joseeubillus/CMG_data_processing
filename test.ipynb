```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# create a plot x and y axis, y function is a sin function
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# plot the function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x)')
plt.legend(['sin(x)'])
plt.show()
```
